```python
from kymatio import Scattering1D
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
import librosa
import librosa.display
import scipy
import torch

%matplotlib inline
```

```python
def _load_audio(path, standart_sr=16000, length = 66000, random_start=False):
    """
    Loading data from *.wav file
    :param path: path to *.wav.file
    :return:
    """
    x, sr = librosa.load(path, sr=None)
    assert sr == 16000
    if length > len(x):
        x = np.concatenate([x] * int(np.ceil(length / len(x))))
    if random_start:
        x = x[random.randint(0, len(x) - length):]
    feature = x[:length]
    return sr, feature / np.max(np.abs(feature))
```

```python
# original
original_filename = "test_spoof_audios/T_1000001.wav"
sample_rate, samples = _load_audio(original_filename)
plt.figure(figsize=(12, 4))
librosa.display.waveplot(samples.astype('float'), sr=sample_rate)
plt.show()

# spoofed
spoofed_filename = "test_spoof_audios/T_1001538.wav"
sample_rate, samples_spoofed = _load_audio(spoofed_filename)
plt.figure(figsize=(12, 4))
librosa.display.waveplot(samples_spoofed.astype('float'), sr=sample_rate)
plt.show()

# original
X = librosa.stft(samples.astype('float'))
Xdb = librosa.amplitude_to_db(X)
plt.figure(figsize=(12, 5))
librosa.display.specshow(Xdb, sr=sample_rate, x_axis='time', y_axis='hz')
plt.show()

# spoofed
X_spoofed = librosa.stft(samples_spoofed.astype('float'))
Xdb_spoofed = librosa.amplitude_to_db(X_spoofed)
plt.figure(figsize=(12, 5))
librosa.display.specshow(Xdb_spoofed, sr=sample_rate, x_axis='time', y_axis='hz')
plt.show()
```

```python
J = 8
Q = 16

samples = np.resize(samples, (1, samples.size))

scattering = Scattering1D(J, samples.shape[1], Q)

# get the metadata on the coordinates of the scattering
meta = Scattering1D.compute_meta_scattering(J, Q)
order0 = (meta['order'] == 0)
order1 = (meta['order'] == 1)
order2 = (meta['order'] == 2)

data_torch = torch.from_numpy(samples)

s = scattering.forward(data_torch)[0]
plt.figure(figsize=(10, 10), dpi=300)
plt.subplot(3, 1, 1)
plt.plot(s[order0].numpy())
plt.title("Scattering order 0")
plt.subplot(3, 1, 2)
plt.imshow(s[order1].numpy(), aspect='auto')
plt.title("Scattering order 1")
plt.subplot(3, 1, 3)
plt.imshow(s[order2].numpy(), aspect='auto')
plt.title("Scattering order 2")

plt.show()
```

```python
samples_spoofed = np.resize(samples_spoofed, (1, samples_spoofed.size))

scattering = Scattering1D(J, samples_spoofed.shape[1], Q)

# get the metadata on the coordinates of the scattering
meta = Scattering1D.compute_meta_scattering(J, Q)
order0 = (meta['order'] == 0)
order1 = (meta['order'] == 1)
order2 = (meta['order'] == 2)

data_torch_spoofed = torch.from_numpy(samples_spoofed)

s = scattering.forward(data_torch_spoofed)[0]
plt.figure(figsize=(10, 10), dpi=300)
plt.subplot(3, 1, 1)
plt.plot(s[order0].numpy())
plt.title("Scattering order 0")
plt.subplot(3, 1, 2)
plt.imshow(s[order1].numpy(), aspect='auto')
plt.title("Scattering order 1")
plt.subplot(3, 1, 3)
plt.imshow(s[order2].numpy(), aspect='auto')
plt.title("Scattering order 2")

print(s[order2].numpy().shape)

plt.show()
```

```python
import pandas as pd

J = 8
Q = 12
T = 48000
log_eps = 1e-6
torch.manual_seed(42)

def read_files(train=True):
    filename = 'audio_list_train.txt'
    path = 'ASVspoof2017_V2_train/'
    if(not train):
        filename = 'audio_list_dev.txt'
        path = 'ASVspoof2017_V2_dev/'
        
    files = open(filename).readlines()
    x_all = torch.zeros(len(files), T, dtype=torch.float32)
    y_all = torch.zeros(len(files), dtype=torch.int64)
    for k, line in enumerate(files):
        filename, y, dictor, frase_id, _, _, _ = line.strip().split()

        if(y == 'spoof'):
            y = 1
        else:
            y = 0
        
        # Load the audio signal and normalize it.
        _, x = wavfile.read(path + filename)
        x = np.asarray(x, dtype='float')
        x /= np.max(np.abs(x))

        # Convert from NumPy array to PyTorch Tensor.
        x = torch.from_numpy(x)

        # If it's too long, truncate it.
        if x.numel() > T:
            x = x[:T]

        # If it's too short, zero-pad it.
        start = (T - x.numel()) // 2

        x_all[k,start:start + x.numel()] = x
        y_all[k] = y
    
    return x_all, y_all
        
X, y = read_files()
```

```python
print(X.shape, y.shape)
```

```python
scattering = Scattering1D(J, T, Q)
Sx_all = scattering.forward(X)
```

```python
log_eps = 1e-6
Sx_all = torch.log(torch.abs(Sx_all) + log_eps)
```

```python
Sx_all = torch.mean(Sx_all, dim=-1)
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.multiclass import OneVsRestClassifier
```

```python
pca = PCA(n_components=4)
pca_result = pca.fit_transform(Sx_all)
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
plt.scatter(pca_result[:, 1], pca_result[:, 3],
            c=y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('brg', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
plt.show();
```

```python
# проверяем на тестовой выборке
X_dev, y_dev = read_files(train=False)
print(X_dev.shape, y_dev.shape)

Sx_dev_all = scattering.forward(X_dev)
Sx_dev_all = torch.log(torch.abs(Sx_dev_all) + log_eps)
Sx_dev_all = torch.mean(Sx_dev_all, dim=-1)
```

```python
pca_result_dev = pca.transform(Sx_dev_all)
print(pca_result_dev)

plt.scatter(pca_result_dev[:, 1], pca_result_dev[:, 3],
            c=y_dev, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('brg', 10))
plt.xlabel('component 1 test')
plt.ylabel('component 2 test')
plt.colorbar();
plt.show();
```

```python
import time

data_clone = Sx_all.clone()
time_start = time.time()
tsne = TSNE(n_components=2, verbose=1, perplexity=30, n_iter=300)
tsne_results = tsne.fit_transform(data_clone)

print('t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

```python
plt.scatter(tsne_results[:, 0], tsne_results[:, 1],
            c=y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('brg', 10))
plt.xlabel('tsne-2d-one')
plt.ylabel('tsne-2d-two')
plt.colorbar();
plt.show();
```

```python
# проверяем тестовую
data_clone = Sx_dev_all.clone()
time_start = time.time()
tsne = TSNE(n_components=2, verbose=1, perplexity=30, n_iter=300)
tsne_dev_results = tsne.fit_transform(data_clone)

print('t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

```python
plt.scatter(tsne_dev_results[:, 0], tsne_dev_results[:, 1],
            c=y_dev, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('brg', 10))
plt.xlabel('tsne-2d-one test')
plt.ylabel('tsne-2d-two test')
plt.colorbar();
plt.show();
```
