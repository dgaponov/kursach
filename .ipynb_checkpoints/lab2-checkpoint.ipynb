```python
from skimage import io,color
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import os
import scipy.misc
from sklearn import decomposition
%matplotlib inline
```

```python
image_name = "tulen.jpg"
img_gray = io.imread(image_name, as_gray=True)
io.imshow(img_gray)
xlabel('Original Image')
show()
original_size = os.path.getsize(image_name)
print(original_size)
print(img_gray.shape)
```

```python
components = [1, 5, 10, 20, 50, 100, 125, 250, 400]
min_varience = 100

image_sizes = []

plt.figure(1)

for num_comp in components:
    pca = decomposition.PCA(num_comp)
    Y_sklearn = pca.fit_transform(img_gray)
    inv_tr = pca.inverse_transform(Y_sklearn)
    
    path = "result_" + str(num_comp) + ".jpg"
    scipy.misc.imsave(path, inv_tr)
    
    img = mpimg.imread(path)
    plt.xlabel("n_components = " + str(num_comp))
    plt.imshow(img, cmap="gray")
    plt.show()
    
    image_sizes.append(os.path.getsize(path))
    
    print(Y_sklearn.shape)
    print(len(pca.explained_variance_ratio_), pca.explained_variance_ratio_[-1])
    
    
    if(pca.explained_variance_ratio_[-1] < min_varience):
        min_varience = pca.explained_variance_ratio_[-1]

print(min_varience)
```

```python
print(components)
print(image_sizes)
print("original", original_size)
compress = [original_size * 100 / image_size for image_size in image_sizes]
print(compress)
```

```python
plt.plot(components, compress, 'ro-')
plt.grid(True)
plt.ylabel(u'Степень сжатия в %')
plt.xlabel(u'Количество главных компонент')
plt.show()
```
